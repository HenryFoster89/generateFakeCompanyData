```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"data_output\Ordinato.csv")

# 1) Ensure datetime
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

for mat in df["MaterialID"].unique():
    dfTemp = df.loc[df["MaterialID"] == mat, ["Date", "QuantityOrdered"]].dropna(subset=["Date"])
    dfTemp = dfTemp.sort_values("Date")

    # 2) Monthly aggregation (month start)
    s = dfTemp.groupby(pd.Grouper(key="Date", freq="MS"))["QuantityOrdered"].sum()

    # 3) Plot: s is a Series (index = Date)
    fig, ax = plt.subplots(figsize=(14, 10))
    ax.plot(s.index, s.values)

    ax.set_title(f"Demand time series - MaterialID: {mat}")
    ax.set_xlabel("Date")
    ax.set_ylabel("QuantityOrdered")

    plt.show()
```
